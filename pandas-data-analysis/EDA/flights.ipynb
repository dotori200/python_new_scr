```python
# 연도별 승객 수의 변화 lineplot - sns
import matplotlib.pyplot as plt
import seaborn as sns
sns.lineplot (data = flights, x='year' , y='passengers' , errorbar = ('ci' , 95))  # errorbar : 신뢰구간 보여줌
flights.pivot(index = 'year', columns = 'month', values = 'passengers')
# seaborn 시각화는 신뢰 구간도 같이 표시하는 경우
# 데이터가 변동될 수 있는 구간
# 신뢰구간 넓으면 --> 불확실성이 크다 (데이터 분산이 크거나, 샘플 수가 작은 경우)
# 신뢰구간 좁으면 --> 안정적이 데이터, 정확성이 비교적 높다 (분산이 작거나(데이터가 집중적으로 모여있거나), 샘플 수가 많은 경우)
plt.show()

```
