```python
# 결측치 : 다양한 이유로 불완전한 데이터 - 누락 None NaN

import pandas as pd
import numpy as np
list (np.array( [1,2,3])) 

df = pd.DataFrame ({

    'A' : [1,2,np.nan,4],
    'B' : [5,np.nan,np.nan,8],
    'C' : [9,10,11,12]
})

print ('결측치 확인(isna)')
df
```

```python
# 결측치 제거 
# dropna() 결측치가 포함된 행 또는 열을 제거 axis = 0 기본값 / how = 'any' : 결측치가 하나라도 있을 시 해당 행 전체를 제거
# 열 제거하려면 뒤에 (axis = 1) 붙여주기
# how = 'all' 모든 값이 결측치일 떄 삭제 thresh : 지정한 개수 이상의 비결측치 값이 있는 행 , 열을 유지
df.dropna (axis = 1) 

```

```python
# 결측치 대처방법
# fillna() : 괄호 안의 값으로 결측치를 채움
df.fillna(0)
print (f'각 컬럼의 평균값 \n {df.mean()}')

# 수치데이터가 아닌 object형은 df.mode()최빈값
df.fillna(method='ffill') # bfill도 가능
```
